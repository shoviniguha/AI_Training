```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Product": ["Laptop", "Laptop", "Laptop", "Mobile", "Mobile", "Mobile"],
    "Units_Sold": [120, 135, 150, 200, 220, 210],
    "Revenue": [600000, 675000, 750000, 300000, 330000, 315000]
}
 
df = pd.DataFrame(data)
df
```

```python
total_revenue=df.groupby('Product')['Revenue'].sum()

total_revenue.plot(kind='bar',color='skyblue',edgecolor='red')
plt.title("Total Revenue by Product- Shovini")
plt.xlabel('Product')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
total_sales=df.groupby('Product')['Units_Sold'].sum()

total_sales.plot(kind='bar',color='red',edgecolor='blue')
plt.title("Total Sales by Product- Shovini")
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.show()
```
