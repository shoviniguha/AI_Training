```python
import pandas as pd
import matplotlib.pyplot as plt
# Sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
df = pd.DataFrame(data)
df
```

```python
#TASK 1: Basic Analysis
```

```python
total_revenue=df.groupby('Product')['Revenue'].sum()
average_units=df.groupby('Product')['Units_Sold'].mean()
```

```python
print(total_revenue)
print(average_units)
```

```python
#TASK 2: Visualization 1 â€“ Line Chart
```

```python
monthly_revenue=df.groupby(['Month','Product'])['Revenue'].sum()
```

```python
monthly_revenue
```

```python
import numpy as np
monthly_revenue=np.array(monthly_revenue)
monthly_revenue
```

```python
laptop=[]
for i in (3,0,6):
    laptop.append(monthly_revenue[i])
mobile=[]
for i in (4,1,7):
    mobile.append(monthly_revenue[i])
tv=[]
for i in (5,2,8):
    tv.append(monthly_revenue[i])
```

```python
months=['Jan','Feb','Mar']
```

```python
import matplotlib.pyplot as plt
plt.plot(months,laptop,label='Laptop',marker='o')
plt.plot(months,mobile,label='Mobile',marker='o')
plt.plot(months,tv,label='TV',marker='o')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=30)
plt.legend()
plt.title("Monthly Revenue Trend for each product")
plt.show()
```
